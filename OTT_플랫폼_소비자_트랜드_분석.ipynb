```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict

# 파일 읽기
with open('genre_restored.csv', encoding='cp949', errors='ignore') as fp:
    data_genre = csv.reader(fp)

    genre = []
    age = []
    country = []

    # 데이터를 순회하며 연령대와 장르, 국가를 수집
    for row in data_genre:
        if '연령별' in row[3]:  # '연령별'이 포함된 행만 필터링
            genre.append(row[6:])  # 장르 리스트 저장 (장르 정보는 6번째 이후 열)
            age.append(row[4])     # 연령대 저장 (4번째 열)
            country.append(row[2]) # 국가 저장 (2번째 열)

# 국가별 연령대별 선호 장르를 시각화할 수 있도록 데이터 준비
country_age_genre = defaultdict(lambda: defaultdict(list))

# 데이터를 국가별로 연령대별로 장르를 저장
for i in range(len(country)):
    current_country = country[i]
    current_age = age[i]
    current_genre = genre[i]

    # 연령대별로 국가의 장르 데이터를 누적
    country_age_genre[current_country][current_age].extend(current_genre)

# 장르별로 데이터를 카운팅하기 위한 구조
countries = list(country_age_genre.keys())  # 국가 리스트 생성
genres = list(set([g for sublist in genre for g in sublist]))  # 장르 리스트 생성

# 각 국가별로 장르 카운트를 저장
genre_counts_by_country = {country: {genre: 0 for genre in genres} for country in countries}

# 데이터 카운팅
for country in countries:
    for age in country_age_genre[country]:
        for genre in country_age_genre[country][age]:
            genre = genre.strip()  # 공백 제거
            if genre in genre_counts_by_country[country]:
                genre_counts_by_country[country][genre] += 1

# 상위 5개 장르만 추출하는 로직 추가
top_n = 5
top_genres_by_country = {}

for country in countries:
    sorted_genres = sorted(genre_counts_by_country[country].items(), key=lambda x: x[1], reverse=True)
    top_genres_by_country[country] = sorted_genres[:top_n]

# 시각화 (국가별 상위 5개 장르만 막대 그래프로 비교)
bar_width = 0.2
index = np.arange(top_n)  # 상위 5개의 장르에 대한 인덱스

plt.figure(figsize=(12, 8))

for i, country in enumerate(countries):
    top_genres = [item[0] for item in top_genres_by_country[country]]
    top_counts = [item[1] for item in top_genres_by_country[country]]

    plt.bar(index + i * bar_width, top_counts, bar_width, label=country)

# X축과 Y축 설정
plt.xlabel('장르')
plt.ylabel('선호도')
plt.title('국가별 상위 5개 선호 장르')
plt.xticks(index + bar_width, top_genres_by_country[countries[0]], rotation=90)  # 첫 번째 국가의 장르 이름을 사용

# 범례 추가
plt.legend()

# 그래프 출력
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('genre_restored.csv', encoding='cp949')

# Initialize dictionaries to store male and female population by country
country_male_population = {}
country_female_population = {}

# Iterating over rows in the dataframe
for index, row in df.iterrows():
    country = row['국가']  # '국가' column contains the country information
    population = row['사례수(명)']  # '사례수(명)' column contains the population number
    gender = row['구분']  # '구분' column contains gender (Male or Female)

    # Checking gender and adding to the respective dictionary
    if gender == '남성':  # If the gender is Male
        if country in country_male_population:
            country_male_population[country] += population
        else:
            country_male_population[country] = population
    elif gender == '여성':  # If the gender is Female
        if country in country_female_population:
            country_female_population[country] += population
        else:
            country_female_population[country] = population

# Output the results
print("Country-wise Male Population:", country_male_population)
print("Country-wise Female Population:", country_female_population)

# Plotting the population by gender for each country
countries = list(country_male_population.keys())
male_populations = list(country_male_population.values())
female_populations = [country_female_population.get(country, 0) for country in countries]

# Plotting the results
width = 0.4
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for male population
ax.bar(countries, male_populations, width, label='Male', color='b', align='center')
# Bar plot for female population
ax.bar(countries, female_populations, width, label='Female', color='r', align='edge')

# Adding labels and title
ax.set_xlabel('Country')
ax.set_ylabel('Population')
ax.set_title('Male and Female Population by Country')
ax.legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('genre_restored.csv', encoding='cp949')

# Initialize dictionaries for education and income populations
education_population = {}
income_population = {}

# Iterating over rows in the dataframe
for index, row in df.iterrows():
    education = row['구분']
    income = row['구분']
    population = row['사례수(명)']

    # Processing education data
    if education in ['고졸 이하', '대졸', '대졸 이상', '대학원 이상']:
        if education in education_population:
            education_population[education] += population
        else:
            education_population[education] = population

    # Processing income data
    if income in ['40,000 이상', '40,000 미만', '8,000 이상', '8,000 미만', '5,000 이상', '5,000 미만']:
        if income in income_population:
            income_population[income] += population
        else:
            income_population[income] = population

# Data preparation
education_levels = list(education_population.keys())
income_levels = list(income_population.keys())

education_populations = list(education_population.values())
income_populations = list(income_population.values())

# Create figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot income as a bar chart
width = 0.4
x_income = range(len(income_levels))
ax1.bar(x_income, income_populations, width, label='Income', color='orange', align='center')

# Create another axis for the line plot (education)
ax2 = ax1.twinx()  # Create a second y-axis sharing the same x-axis

# Plot education data as a line chart
x_education = range(len(education_levels))
ax2.plot(x_education, education_populations, marker='o', label='Education', color='blue', linestyle='-', linewidth=2)

# Set X-axis labels and ticks
ax1.set_xticks(list(range(len(income_levels))))
ax1.set_xticklabels(income_levels, rotation=45)

# Adding labels and titles
ax1.set_xlabel('Income Levels (Bar) / Education Levels (Line)')
ax1.set_ylabel('Population (Income)', color='orange')
ax2.set_ylabel('Population (Education)', color='blue')

# Add title
plt.title('Population by Income (Bar) and Education (Line)')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('genre_restored.csv', encoding='cp949')

# 직업별로 필터링
occupations = ['사무·전문직', '서비스·생산직', '학생/주부/무직', '농림어업/군인', '기타/무직']
# 소득별로 필터링
income_levels = ['40,000 이상', '40,000 미만', '8,000 이상', '8,000 미만', '5,000 이상', '5,000 미만']

# Initialize a dictionary to store occupation and income data
occupation_income_population = {occupation: {income: 0 for income in income_levels} for occupation in occupations}

# Populate the dictionary with data from the dataframe
for index, row in df.iterrows():
    occupation = row['구분']
    income = row['구분']
    population = row['사례수(명)']

    # 직업과 소득에 따라 데이터를 저장
    if occupation in occupations and income in income_levels:
        occupation_income_population[occupation][income] += population

# Prepare data for the stacked bar chart
occupation_levels = list(occupation_income_population.keys())

# Create lists for each income level
data_for_income_levels = {income: [] for income in income_levels}

# Fill in the data for each income level
for occupation in occupation_levels:
    for income in income_levels:
        data_for_income_levels[income].append(occupation_income_population[occupation][income])

# Plot the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each income level's data as a stacked bar chart
bottom = [0] * len(occupation_levels)  # Initialize the bottom position for the stacked bars
for income in income_levels:
    ax.bar(occupation_levels, data_for_income_levels[income], label=income, bottom=bottom)
    bottom = [i + j for i, j in zip(bottom, data_for_income_levels[income])]  # Update the bottom for the next stack

# Add labels and title
ax.set_xlabel('Occupation')
ax.set_ylabel('Population')
ax.set_title('Stacked Population by Occupation and Income')

# Add a legend
ax.legend(title='Income Levels')

# Display the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('genre_restored.csv', encoding='cp949')

# 직업과 소득 데이터
occupations = ['사무·전문직', '서비스·생산직', '학생/주부/무직', '농림어업/군인', '기타/무직']
income_levels = ['40,000 이상', '40,000 미만', '8,000 이상', '8,000 미만', '5,000 이상', '5,000 미만']

# Initialize a dictionary to store occupation and income data
occupation_income_population = {occupation: {income: 0 for income in income_levels} for occupation in occupations}

current_occupation = None  # 직업을 추적하기 위한 변수

# 직업과 소득을 연결하여 데이터를 매핑
for index, row in df.iterrows():
    if row['구분'] in occupations:  # 직업을 먼저 찾음
        current_occupation = row['구분']
    elif row['구분'] in income_levels and current_occupation:  # 해당 직업에 대한 소득 정보를 추가
        income = row['구분']
        population = row['사례수(명)']
        occupation_income_population[current_occupation][income] += population

# Prepare data for the stacked bar chart
occupation_levels = list(occupation_income_population.keys())

# Create lists for each income level
data_for_income_levels = {income: [] for income in income_levels}

# Fill in the data for each income level
for occupation in occupation_levels:
    for income in income_levels:
        data_for_income_levels[income].append(occupation_income_population[occupation][income])

# Plot the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each income level's data as a stacked bar chart
bottom = [0] * len(occupation_levels)  # Initialize the bottom position for the stacked bars
for income in income_levels:
    ax.bar(occupation_levels, data_for_income_levels[income], label=income, bottom=bottom)
    bottom = [i + j for i, j in zip(bottom, data_for_income_levels[income])]  # Update the bottom for the next stack

# Add labels and title
ax.set_xlabel('Occupation')
ax.set_ylabel('Population')
ax.set_title('Stacked Population by Occupation and Income')

# Add a legend
ax.legend(title='Income Levels')

# Display the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
